```python
import json
import requests
from datetime import datetime, timedelta
import pandas as pd


df_all = pd.DataFrame()

start_date = datetime(2020, 2, 1)
end_date = datetime(2020, 12, 31)
for i in range(0, 365):
    current_date = start_date + timedelta(days=i)
    date_string = current_date.strftime("%Y-%m-%d")
    querystring = {"city_name":"Autauga","region_province":"Alabama","iso":"USA","region_name":"US","q":"US Alabama","date":date_string}
    response = requests.request("GET", url, headers=headers, params=querystring)
    data = response.json()
    df = pd.DataFrame(data['data'], index=[0])
    df_all = pd.concat([df_all, df])
```

```python
df_all2 = df_all[df_all.date.notnull()]
x = df_all2[df_all2.columns[-12:-1]]
x
```

```python
x["month"] = pd.PeriodIndex(x["date"],freq = "M")
x
```

```python
#Ordenamos los datos 
x2 = x.sort_values('date').groupby('month').last().reset_index()
x2
```

```python
#Grafico de barras
import matplotlib.pyplot as plt
x2.plot(kind = "bar", x = "month", y = "deaths")
plt.title("de Marzo a Enero, en millones")
plt.suptitle("Evolución de muertes por mes 2020-2021")
"Evolución de muertes por mes"
plt.show()
```

### Datos estadísticos

```python
#Muertes
for i in range(0,11):
    if i <=10:
        xd=x2.loc[i, 'deaths']
        dx=x2.loc[i+1, 'deaths']
        print((((dx-xd)/xd)*100).round(2),'%')
```

#### Observamos que en el caso de USA hubo un gran salto en el crecimiento de la tasa de muertes durante los primeros meses siendo el mayor  387200% y los demas con menores variaciones que oscilan entre 10% a 30%.

```python
#Casos confirmados
for i in range(0,11):
    if i <=10:
        xd=x2.loc[i, 'confirmed']
        dx=x2.loc[i+1, 'confirmed']
        print((((dx-xd)/xd)*100).round(2),'%')
```

#### Con los casos cofirmados  es algo parecido  con un gran crecimiento en los primeros meses siendo el mayor un 268717.14% , con respecto a los demas datos su crecimiento varia un poco mas oscilando entre 20% a 70%.

```python
#Casos activos
for i in range(0,11):
    if i <=10:
        xd=x2.loc[i, 'active']
        dx=x2.loc[i+1, 'active']
        print((((dx-xd)/xd)*100).round(2),'%')
```

#### Siguiendo el patron, en los primeros mese hay un crecimiento elevado siendo el mayor  de 285827.42%, y con los demas datos a diferencia de las anteriores observaciones aca hubo un crecimiento fuera de lo común pasando de un 53.85% a 141.45%.
