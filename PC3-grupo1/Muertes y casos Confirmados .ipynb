#### 1) Muertes y Casos Confirmados por país hasta el 2022-12-31 por Covid-19:

```python


url = "https://covid-19-statistics.p.rapidapi.com/regions"#Guardamos el link de la appi de todos los países
URL= "https://covid-19-statistics.p.rapidapi.com/reports/total" #Guardamos el link de la appi de los reportes total
headers = {
	"X-RapidAPI-Key": "5715c80c9emsh5051181da9b627fp1d0162jsn4d23a1fc2ead",
	"X-RapidAPI-Host": "covid-19-statistics.p.rapidapi.com"
}

response = requests.request("GET", url, headers=headers)#Leemos 
df = pd.read_json(response.text)#Convertimos en un dataframe del formato json

col_a = df['data']#leemos el la 1ra columna
a=0        #Añadimos un contador y listas donde almacenaremos 
paises=[]
iso=[]
deaths=[]
casos=[]
for z in range(len(col_a)):#Creamos el bucle con un rango de todos los países
    c=c=col_a[a]           #Guardamos cada elemento
    c=str(c)               #Convertimos en str
    b=re.findall(r"'name':.*'(.*)'",c) #Limpiamos el nombre del país
    paises.append(b)                   #Guardamos el nombre en la lista
    g=re.findall(r"iso.*'(.*)',",c)    #Limpiamos el iso del país
    iso.append(g)                      #guardamos en su lista respectiva
    a=a+1
    querystring ={"iso":g,"date":"2022-12-31"}
    responses = requests.request("GET", URL, headers=headers, params=querystring)
    r=responses.text
    res = ast.literal_eval(r)
    if len(res["data"])!=0:
        muer=res["data"]["deaths"]
        infe=res["data"]["confirmed"]
        deaths.append(muer)
        casos.append(infe)
    else:
        deaths.append("NA")
        casos.append("NA")       
```

```python
datos = {   
    'ISO':iso,             #Creamos un diccionario con las listas 
    'País': paises,
    'Casos': casos,
    'Muertes' : deaths
}

df = pd.DataFrame(datos)# lo convertimos a un dataframe 
df
```
